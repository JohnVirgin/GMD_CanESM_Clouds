```python
## Import Packages
import numpy as np
import _pickle as pk
import AA as aa
from matplotlib import pyplot as plt
import matplotlib as mpl
from netCDF4 import Dataset
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
import matplotlib.ticker as mticker
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

plt.rcParams['axes.xmargin'] = 0 #default margin spacing for the edges
mpl.rc('font',size=40,**{'family':'sans-serif','sans-serif':['Arial']})
%matplotlib inline
```

```python
Source = '/Volumes/iSSD0/Users/johnvirgin/'

cl = np.squeeze(Dataset(Source+'clisccp_CFMon_CanESM2_piControl_r1i1p1_244101-246012.nc')\
                .variables['clisccp'])
cl[cl > 1e10] = np.nan
```

```python
nmon = cl[:, 0, 0, 0, 0].size
nyr = int(nmon/12)
print('number of years read in of picon data - ',nyr)

print('Calculate anomalies')
#convert time dimensions to months/years, then take the climatological mean
cl_stacked = np.zeros([nyr, 12, 7, 7, 64, 128])

s = 0
f = 12

for i in range(nyr):
    cl_stacked[i, :, :, :, :, :] = np.stack(cl[s:f, :, :, :, :], axis=0)

    s += 12
    f += 12
    
climo_cl = np.nanmean(cl_stacked,axis=0)
climo_cl.shape
```

```python
climo_cl_hi_sum = np.tile(\
    np.nansum(climo_cl[:,:,2:,:,:],axis=(1,2))[:,None,None,:,:],(1,7,2,1,1))
climo_cl_lo_n = climo_cl[:,:,:2,:,:]/(100-climo_cl_hi_sum)
climo_cl_lo = np.nansum(climo_cl[:,:,:2,:,:],axis=(1,2))
```

```python
hi = np.nanmean(np.nansum(climo_cl[:,:,2:,:,:],axis=(1,2)),axis=0)
lo_un = np.nanmean(np.nansum(climo_cl_lo_n,axis=(1,2)),axis=0)*100
lo = np.nanmean(np.nansum(climo_cl[:,:,:2,:,:],axis=(1,2)),axis=0)
```

```python
np.amax(lo_un[:,0])
```

```python
lat = np.linspace(-87.864,87.864,64)
lon = np.linspace(0,357.1875,128)
```

```python
hi_plot = np.concatenate((hi[:,64:],hi[:,0:64]),axis=1)
lo_plot = np.concatenate((lo[:,64:],lo[:,0:64]),axis=1)
lo_un_plot = np.concatenate((lo_un[:,64:],lo_un[:,0:64]),axis=1) 

hi_plot,lon_cyc = add_cyclic_point(hi_plot,lon)
lo_plot,lon_cyc = add_cyclic_point(lo_plot,lon)
lo_un_plot,lon_cyc = add_cyclic_point(lo_un_plot,lon)
```

```python
SST_bounds = np.arange(0,100,5)
SST_bounds_ticks = np.arange(0,100,10)
cmap = plt.cm.Blues_r
SST_norm = mpl.colors.BoundaryNorm(SST_bounds, cmap.N)
```

```python
fig = plt.figure(figsize=(30,30),dpi=300)

#Uniform SST change
ax = plt.subplot(321,projection=ccrs.PlateCarree(central_longitude=180))
#ax.set_extent([-100, 30, -60, 60], crs=ccrs.PlateCarree())
ax.set_title('Low Unobscured',fontsize=20,pad=10)
im = ax.contourf(lon_cyc,lat,lo_un_plot,\
                   SST_bounds,cmap=cmap,norm=SST_norm,extend='both')
ax.coastlines()
cbar_1 = plt.colorbar(im,ticks=SST_bounds_ticks,drawedges=True,shrink=0.7)
cbar_1.set_label('%',labelpad=15,fontsize=16)
#plt.text(0.15,1.3,'2SST',fontsize=20,transform = ax.transAxes)

plt.savefig('Lo_un_Cloud.png',bbox_inches='tight')
```
